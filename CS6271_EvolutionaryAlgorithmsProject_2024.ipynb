Set GP parameters.

```python
# Genetic Programming Parameters
POPULATION_SIZE = 2000
P_CROSSOVER = 0.9  
P_MUTATION = 0.15   
MAX_GENERATIONS = 100  
HALL_OF_FAME_SIZE = 15 

RANDOM_SEED = 52
random.seed(RANDOM_SEED)

# Tree Parameters
MIN_TREE_HEIGHT = 2
MAX_TREE_HEIGHT = 5
LIMIT_TREE_HEIGHT = 14
MUT_MIN_TREE_HEIGHT = 3
MUT_MAX_TREE_HEIGHT = 3 
```

```python
from deap import base, creator, tools
# Define a custom fitness function
def fitness_function(individual, X_train, y_train):
    func = toolbox.compile(expr=individual)  # compile GP individual to function
    predictions = [func(*x) for x in X_train]  # apply function to each sample
    accuracy = sum(int(pred == actual) for pred, actual in zip(predictions, y_train)) / len(y_train)
    return accuracy,
```

```python
pset = gp.PrimitiveSet("MAIN", X_train.shape[1])  #Number of input variables

# Define custom operations if necessary
def safe_div(x, y):
    return x / y if y != 0 else 1

def safe_tan(x):
    result = np.tan(x)
    return np.clip(result, -10, 10)  # Clip extreme values

# Add primitives
pset.addPrimitive(np.add, 2)
pset.addPrimitive(np.subtract, 2)
pset.addPrimitive(np.multiply, 2)
pset.addPrimitive(safe_div, 2)   # Assuming safe_div is defined
pset.addPrimitive(np.sin, 1)
pset.addPrimitive(np.cos, 1)
pset.addPrimitive(safe_tan, 1)
# Add terminals (constants)
pset.addEphemeralConstant("rand101", lambda: random.randint(-1, 1))

# Set up individuals and population
creator.create("FitnessMax", base.Fitness, weights=(1.0,))  # maximize fitness
creator.create("Individual", gp.PrimitiveTree, fitness=creator.FitnessMax, pset=pset)

# Create the toolbox
toolbox = base.Toolbox()
toolbox.register("expr", gp.genHalfAndHalf, pset=pset, min_=MIN_TREE_HEIGHT, max_=MAX_TREE_HEIGHT)
toolbox.register("individual", tools.initIterate, creator.Individual, toolbox.expr)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# Register GP functions in the toolbox
toolbox.register("compile", gp.compile, pset=pset)
toolbox.register("evaluate", fitness_function, X_train=X_train, y_train=y_train)
toolbox.register("select", tools.selTournament, tournsize=3) #tournament size
toolbox.register("mate", gp.cxOnePoint)
toolbox.register("expr_mut", gp.genFull, min_=MUT_MIN_TREE_HEIGHT, max_=MUT_MAX_TREE_HEIGHT)
toolbox.register("mutate", gp.mutUniform, expr=toolbox.expr_mut, pset=pset)

# Limit tree height
toolbox.decorate("mate", gp.staticLimit(key=operator.attrgetter("height"), max_value=LIMIT_TREE_HEIGHT))
toolbox.decorate("mutate", gp.staticLimit(key=operator.attrgetter("height"), max_value=LIMIT_TREE_HEIGHT))
```

```python
population = toolbox.population(n=POPULATION_SIZE)
hof = tools.HallOfFame(HALL_OF_FAME_SIZE)

# Statistics to keep track of the best individual
stats = tools.Statistics(lambda ind: ind.fitness.values)
stats.register("avg", np.mean)
stats.register("std", np.std)
stats.register("min", np.min)
stats.register("max", np.max)

# Run the genetic programming algorithm
algorithms.eaSimple(population, toolbox, cxpb=P_CROSSOVER, mutpb=P_MUTATION,
                    ngen=MAX_GENERATIONS, stats=stats, halloffame=hof, verbose=True)

# Define a prediction function for a GP individual
def predict(individual, X):
    func = toolbox.compile(expr=individual)
    return [func(*x) for x in X]

# Use the best individual in the hall of fame to predict the test set
best_individual = hof[0]
y_pred = predict(best_individual, X_test)

with open("submission.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["index", "output"])
    for i, pred in enumerate(y_pred):
        writer.writerow([i, int(pred)])  # Assuming binary predictions are integers 0 or 1      
```
